```r
#Q1
Obj <- c(300,400 )
Constraints <- matrix(c(5,4,3,5,5,4,8,4), nrow = 4, byrow = TRUE)
Inequality <- c("<=","<=",">=", ">=")
Rhs <- c(200,150,100,80)

result <- lp("max", Obj, Constraints, Inequality, Rhs)

cat("Optimal solution:\n")
cat(sprintf("x1\t%.2f\n", result$solution[1]))
cat(sprintf("x2\t%.2f\n", result$solution[2]))

cat("\nOptimal objective value:\n")
cat(sprintf("Z = %.2f\n", result$objval))
```

```r
#Q2
obj <- c(4, 3)
mat <- matrix(c(200, 100,
                1, 2,
                40, 40), nrow = 3, byrow = TRUE)
rhs <- c(4000, 50, 1400)
dir <- rep(">=", 3)
lp_solution <- lp("min", obj, mat, dir, rhs)
if(lp_solution$status == 0) {
  cat("Optimal solution found:\n")
  cat("Minimum cost of the diet:", lp_solution$objval, "\n")
  cat("Units of food A =", lp_solution$solution[1], "\n")
  cat("Units of food B =", lp_solution$solution[2], "\n")
} else {
  cat("No optimal solution found\n")
}
```

```r
#Q3

obj <- c(60, 26, 15, 4.75)
mat <- matrix(c(20, 9, 6, 1,
                10, 4, 2, 1), nrow = 2, byrow = TRUE)
rhs <- c(40, 20)
dir <- c("<=", "<=")
lp_solution <- lp("max", obj, mat, dir, rhs, all.int = TRUE)
if(lp_solution$status == 0) {
  cat("Optimal solution found:\n")
  cat("Objective value (Max Z):", lp_solution$objval, "\n")
  cat("X1 =", lp_solution$solution[1], "\n")
  cat("X2 =", lp_solution$solution[2], "\n")
  cat("X3 =", lp_solution$solution[3], "\n")
  cat("X4 =", lp_solution$solution[4], "\n")
} else {
  cat("No optimal solution found\n")
}
```

```r
#7Q BIG  M

obj <- c(3, 1, 1, 1)
mat <- matrix(c(-2, 2, 1, 0,
                3, 1, 0, 1), nrow = 2, byrow = TRUE)
rhs <- c(4, 6)
con_types <- c("=", "=")
mat_artificial <- cbind(mat, diag(2))
obj_artificial <- c(0, 0, 0, 0, 1, 1) 
lp <- lp("min", obj_artificial, mat_artificial, con_types, rhs, compute.sens = FALSE)
solution <- lp$solution[1:4]  # Extract X1, X2, X3, X4
cat("Optimal Solution:")
cat("\nX1 =", solution[1])
cat("\nX2 =", solution[2])
cat("\nX3 =", solution[3])
cat("\nX4 =", solution[4])
cat("\nOptimal Z value =", lp$objval)  # Optimal Z value
```

```r
#Q11

contractor_bids <- matrix(c(9, 7, 9, 10, 
                            14, 17, 18, 12, 
                            19, 20, 21, 18, 
                            15, 19, 18, 19, 
                            10, 15, 21, 16), ncol = 4, byrow = TRUE)
initial_grant <- 50  # in lakh

total_cost <- apply(contractor_bids, 1, sum)

best_contractor_index <- which.min(total_cost)
best_contractor_name <- paste("Contractor", best_contractor_index)
best_cost <- total_cost[best_contractor_index]

total_time <- apply(contractor_bids, 2, sum)

best_time_contractor_index <- which.min(total_time)
best_time_contractor_name <- paste("Contractor", best_time_contractor_index)
quickest_time <- total_time[best_time_contractor_index]

if (best_cost > initial_grant) {
  supplementary_grant <- best_cost - initial_grant
} else {
  supplementary_grant <- 0
}

unsuccessful_contractor_index <- setdiff(1:5, c(best_contractor_index, best_time_contractor_index))
unsuccessful_contractor_name <- paste("Contractor", unsuccessful_contractor_index)

cat("i) Best way of awarding the repair work:\n")
cat("Award the work to", best_contractor_name, "with a total cost of Rs.", best_cost, "lakh.\n\n")

cat("ii) If necessary to seek supplementary grants:\n")
cat("Seek an additional grant of Rs.", supplementary_grant, "lakh.\n\n")

cat("iii) Contractor who will be unsuccessful in the bid:", unsuccessful_contractor_name, "\n")
```
